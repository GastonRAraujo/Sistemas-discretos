```
import matplotlib.pyplot as plt
import numpy as np
```

```
#Funciones----------------------------------------------------------------------


def heron(a,b,xn,yn):
  xk = 1 + yn - a * (xn**2)
  yk = b*xn

  return xk, yk


def disprey(a,b,xn,yn):
  xk = a * xn*(1-xn)-xn*yn
  yk = (1/b) * xn*yn

  return xk, yk


def cremona(a,xn,yn):
  xk = xn * np.cos(a) - (yn - xn*xn) * np.sin(a)
  yk = xn * np.sin(a) + (yn - xn*xn) * np.cos(a)

  return xk, yk


#Ciclos-------------------------------------------------------------------------
#genero los graficos para realizar rapidamente varias iteraciones

def Heron_Cicle(xi,yi,a_,b_,tf):
  x1 = [xi]*(tf)
  x2 = [yi]*(tf)
  for i in range(tf-1):
    x0 = x1[i]
    y0 = x2[i]
    x1[i+1], x2[i+1] = heron(a_,b_,x0,y0)
  plt.scatter(x1,x2, s = 0.1, marker='.')
  



def Disprey_Cicle(xi,yi,a_,b_,tf):
  x1 = [xi]*(tf)
  x2 = [yi]*(tf)
  for i in range(tf-1):
    x0 = x1[i]
    y0 = x2[i]
    x1[i+1], x2[i+1] = disprey(a_,b_,x0,y0)
  plt.scatter(x1,x2, s = 0.1, marker='.')


def Disprey_Cicle_tirar(xi,yi,a_,b_,tf):
  x1 = [xi]*(tf)
  x2 = [yi]*(tf)
  for i in range(tf-1):
    x0 = x1[i]
    y0 = x2[i]
    x1[i+1], x2[i+1] = disprey(a_,b_,x0,y0)
  return x1[-1], x2[-1]

def Cremona_Cicle(xi,yi,a_,tf):
  x1 = [xi]*(tf)
  x2 = [yi]*(tf)
  for i in range(tf-1):
    x0 = x1[i]
    y0 = x2[i]
    x1[i+1], x2[i+1] = cremona(a_,x0,y0)
  plt.scatter(x1,x2, s = 0.1, marker='.')
```

Heron




```
#Parametros -----------------------------
a = 1.4
b = 0.3


t0 = 200


#Memoria ---------------------------------
#Las primeras iteraciones seran descartadas
x = [0]*t0
y = [0]*t0


#Iteraciones iniciales-------------------------

for i in range(t0-1):
  x0 = x[i]
  y0 = y[i]
  x[i+1], y[i+1] = heron(a,b,x0,y0)


Heron_Cicle(x[-1],y[-1],a,b,3000)

plt.xlim(-1.3,1.3)
plt.ylim(-0.5,0.5)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Heron a = 1.4 b = 0.3")
plt.show()



Heron_Cicle(x[-1],y[-1],a,b,20000)

plt.xlim(0.5,0.75)
plt.ylim(0.15,0.21)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Heron a = 1.4 b = 0.3, Cantor Set")
plt.show()
```
