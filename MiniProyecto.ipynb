```
import matplotlib.pyplot as plt
import numpy as np
```

```
#Funciones----------------------------------------------------------------------


def henon(a,b,xn,yn):
  xk = 1 + yn - a * (xn**2)
  yk = b*xn

  return xk, yk


def disprey(a,b,xn,yn):
  xk = a * xn*(1-xn)-xn*yn
  yk = (1/b) * xn*yn

  return xk, yk


def cremona(a,xn,yn):
  xk = xn * np.cos(a) - (yn - xn*xn) * np.sin(a)
  yk = xn * np.sin(a) + (yn - xn*xn) * np.cos(a)

  return xk, yk


#Ciclos-------------------------------------------------------------------------
#genero los graficos para realizar rapidamente varias iteraciones

def Henon_Cicle(xi,yi,a_,b_,tf):
  x1 = [xi]*(tf)
  x2 = [yi]*(tf)
  for i in range(tf-1):
    x0 = x1[i]
    y0 = x2[i]
    x1[i+1], x2[i+1] = henon(a_,b_,x0,y0)
  plt.figure(dpi=300)
  plt.scatter(x1,x2, s = 0.1, marker='.')
  



def Disprey_Cicle(xi,yi,a_,b_,tf):
  x1 = [xi]*(tf)
  x2 = [yi]*(tf)
  for i in range(tf-1):
    x0 = x1[i]
    y0 = x2[i]
    x1[i+1], x2[i+1] = disprey(a_,b_,x0,y0)
  return x1, x2


def Disprey_Cicle_tirar(xi,yi,a_,b_,tf):
  x1 = [xi]*(tf)
  x2 = [yi]*(tf)
  for i in range(tf-1):
    x0 = x1[i]
    y0 = x2[i]
    x1[i+1], x2[i+1] = disprey(a_,b_,x0,y0)
  return x1[-1], x2[-1]

def Cremona_Cicle(xi,yi,a_,tf):
  x1 = [xi]*(tf)
  x2 = [yi]*(tf)
  for i in range(tf-1):
    x0 = x1[i]
    y0 = x2[i]
    x1[i+1], x2[i+1] = cremona(a_,x0,y0)
  plt.scatter(x1,x2, s = 0.1, marker='.')
```

Hénon semicontinuo

```
# import third-party libraries
import numpy as np 
import matplotlib.pyplot as plt

def henon_attractor(x, y, a=.1, b=0.03):
	'''Computes the next step in the Henon 
	map for arguments x, y with kwargs a and
	b as constants.
	'''
	dx = 1 - a * x ** 2 + y
	dy = b * x
	return dx, dy
	
# number of iterations and step size
steps = 5000000
delta_t = 0.0047

X = np.zeros(steps + 1)
Y = np.zeros(steps + 1)

# starting point
X[0], Y[0] = 1, 1

# compute using Euler's formula
for i in range(steps):
	x_dot, y_dot = henon_attractor(X[i], Y[i])
	X[i+1] = X[i] + x_dot * delta_t
	Y[i+1] = Y[i] + y_dot * delta_t

```

```
import matplotlib.patches as patches

# display plot
plt.figure(dpi=300)
plt.plot(X, Y, ',', color='black', alpha = 0.2, markersize=0.013)
plt.axis('on')
plt.ylim(bottom=0.38*1000000)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Héron, a=.1 b=0.03")

# Create a Rectangle patch
rect = patches.Rectangle((-1000, 0.665*1000000), 3800, 0.135*1000000, linewidth=0.5, edgecolor='r', facecolor='none')
0.665*1000000,0.8*1000000
# Add the patch to the Axes
ax = plt.gca()
ax.add_patch(rect)
plt.show()

```
